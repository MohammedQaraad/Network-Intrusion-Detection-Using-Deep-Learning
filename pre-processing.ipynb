```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

file="all"
dir="datasets/multiclass_test"
df = pd.read_csv(f"{dir}/{file}.csv")

#---------performing label encoding to a certain coloum
label_encoder = LabelEncoder() 
scaler=MinMaxScaler()


col='service'
df[col]= label_encoder.fit_transform(df[col]) 

col='flag'
df[col]= label_encoder.fit_transform(df[col]) 

df
```

```python
dummies = pd.get_dummies(df.attack_types)   #contains hot encoded col-1
dummies2 = pd.get_dummies(df.protocol_type) #contains hot encoded col-2
dummies

# Remove column name 'attack_types' 
removed = df.drop(["attack_types"], axis = 1) #removing already encoded col
df_dummies= pd.concat([removed,dummies],axis='columns')

df3=pd.DataFrame(df_dummies)
x1=df.drop(["attack_types"], axis = 1)
x1= pd.concat([x1,dummies],axis='columns')
#--------------------------------------------------
x1=x1.drop(["protocol_type"],axis='columns')
x1=pd.concat([dummies2,x1],axis='columns')
#dummies2
#--------------------------------------------------

scaler=MinMaxScaler()
scaler.fit(x1)
scaled_data=scaler.transform(x1)
scaled_data=pd.DataFrame(scaled_data)
scaled_data.columns= x1.columns

scaled_data
```

```python
scaled_data.to_csv(f"{dir}/processed/{file}.csv", index=False)
```
